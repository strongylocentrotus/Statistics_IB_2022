````python
def demonstrate_clt(expressions, sample_size = 1000, n_samples = 1000):
    mean_expressions = []
    for _ in rp(None, n_samples):
        sample = np.random.choice(expressions, size = sample_size)
        mean_expressions.append(np.mean(sample))
    mean_mean = np.array(mean_expressions).mean()
    return mean_mean
````

````python
b_gene = b_cells_expression_data[example_gene]
nk_gene = nk_cells_expression_data[example_gene]

b_mean = demonstrate_clt(b_gene)
nk_mean = demonstrate_clt(nk_gene)
b_mean, nk_mean
````

А теперь посчитайте 95% доверительные интервалы для обоих распределений (примем тот факт, что средние распределены нормально для обоих клеточных типов) и скажите, отличается ли средняя экспрессия данного гена между клеточными типами?

````python
# для B-клеток
def calculate_ci(sample, mean_mean, std = None):
    if std is None:
      std = sample.std()

    se = std / np.sqrt(len(sample))
    left_b = mean_mean - 1.96 * se
    right_b = mean_mean + 1.96 * se

    return left_b, right_b

b_intervals = calculate_ci(b_gene, b_mean)

b_intervals, b_mean

#NK intervals
nk_intervals = calculate_ci(nk_gene, nk_mean)

nk_intervals, nk_mean
````

````python
def check_intervals_intersect(first_ci, second_ci):
    are_intersect = False
    if  (first_ci[0] < second_ci[1]) and (first_ci[1] > second_ci[0]):
        are_intersect = True
    return are_intersect
````

````python
def check_dge_with_ci(first_table, second_table):
    first_intervals = []
    second_intervals = []
    gen_names = first_table.columns[:-1].values
    for gene in gen_names:
        first_length = len(first_table[gene])
        first_inter = st.t.interval(confidence=0.95,
                        df=first_length - 1,
                        loc=np.mean(first_table),
                        scale=st.sem(first_table))
        first_intervals.append(first_inter)
        second_length = len(second_table[gene])
        second_int = st.t.interval(confidence=0.95,
                        df=second_length - 1,
                        loc=np.mean(second_table),
                        scale=st.sem(second_table))
        second_intervals.append(second_int)
    ci_test_results = []
    for f_int, s_int in zip(
        first_intervals, second_intervals
    ):
        ci_test_results.append(check_intervals_intersect(f_int, s_int))
    return ci_test_results
    
````

````python
def check_dge_with_ztest(first_table, second_table):
    gene_names = first_table.columns[:-1].values
    z_stat_val = []
    for gene in gene_names:
        ans = ztest(
            first_table[gene],
            second_table[gene]
        )
        z_stat_val.append(ans)
    z_test_p_values = []
    for num in z_stat_val:
        z_test_p_values.append(num[1])
    z_test_results = []
    for val in z_test_p_values:
        if val < 0.05:
            z_test_results.append(True)
        else:
            z_test_results.append(False)
    return z_test_results

check_dge_with_ztest(b_cells_expression_data, nk_cells_expression_data)
````
